```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=3, shuffle=True, random_state=42)

print(X[:5, 0])
print(X[:5, 1])
print(y[:5])
```

```python
plt.scatter(X[:, 0], X[:, 1], color='green', marker='o')
plt.grid()
plt.show()
```

```python
clf = KMeans(n_clusters=3, init='random', max_iter=300, tol=0.0004, random_state=42)
y_km = clf.fit_predict(X)
centroids = clf.cluster_centers_
print(centroids)
```

```python
for i in range(len(y_km)):
    index = y_km[i]
    if index == 0:
        plt.scatter(X[i, 0], X[i, 1], color='green', marker='.')
    elif index == 1:
        plt.scatter(X[i, 0], X[i, 1], color='blue', marker='.')
    elif index == 2:
        plt.scatter(X[i, 0], X[i, 1], color='red', marker='.')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', color='orange')
plt.title('KMeans Clustering on random blobs')
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
# Find the optimal number of clusters
# using the elbow method
distortions = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(X)
    distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o', color='black')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('Ideal N Clusters Using Elbow Optimization')
plt.show()
```
