```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate random data to make some blobs
X, y = make_blobs(n_samples=150, n_features=2, centers=3, shuffle=True, random_state=42)

print('5 BLobs x value:', X[:5, 0])
print('5 Blobs y value:', X[:5, 1])
print('5 Blobs Labels:', y[:5])
```

```python
# Graph all current data points
plt.scatter(X[:, 0], X[:, 1], color='green', marker='o')
plt.grid()
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('A graphof all data points')
plt.show()
```

```python
# Create a KMeans classifier and fit data
clf = KMeans(n_clusters=3, init='random', max_iter=300, tol=0.0004, random_state=42)
y_km = clf.fit_predict(X)
centroids = clf.cluster_centers_
print('Centroids:', centroids)
```

```python
# Visualize data through an unsupervised learning lens.
# Graph each data point with a color relating to its class
# Graph each centroid as an orange dot
plt.scatter(X[y_km==0, 0], X[y_km==0, 1], color='green', marker='.', label='Class 0')
plt.scatter(X[y_km==1, 0], X[y_km==1, 1], color='blue', marker='.', label='Class 1')    
plt.scatter(X[y_km==2, 0], X[y_km==2, 1], color='red', marker='.', label='Class 2')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', color='orange')
plt.title('KMeans Clustering on Random Blobs')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()
plt.grid()
plt.show()
```

```python
# Find the optimal number of clusters using the elbow method
# I used n = 3 clusters, and the elbow seems to ve optimal at n = 3
distortions = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(X)
    distortions.append(km.inertia_)

plt.plot(range(1,11), distortions, marker='o', color='black')
plt.xlabel('N clusters')
plt.ylabel('Distortion')
plt.title('Ideal N Clusters Using Elbow Optimization')
plt.show()
```
