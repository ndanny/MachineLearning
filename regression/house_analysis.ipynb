```python
# This notebook is a visual and computational analysis
# of a housing dataset through use of regression models.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset into a dataframe and define column names
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data', header=None, sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
# Visualize some characteristics and their relationships
sns.set(style='whitegrid', context='notebook')
cols = ['LSTAT', 'CRIM', 'TAX', 'RM', 'MEDV']
sns.pairplot(df[cols], size=2.5)
plt.show()
```

```python
# Create a correlation matrix to measure linear dependence
matrix = np.corrcoef(df[cols].values.T)

# Visualize matrix using a heatmap
sns.set(font_scale=1.5)
hm = sns.heatmap(matrix, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 15}, yticklabels=cols, xticklabels=cols)
plt.show()
```

```python
# Implement a regression classifier for RM vs MEDV and train the model
X = df[['RM']].values
y = df['MEDV'].values
clf = LinearRegression()
clf.fit(X, y)
```

```python
# Visualize the the linear model (y = Wx + b)
prediction = clf.predict(X)

plt.scatter(X, y, marker='.', color='purple')
plt.plot(X, prediction, color='orange')
plt.xlabel('Average Number of Rooms [RM]')
plt.ylabel('Median Home Value [MEDV]')
plt.show()
```

```python
# Predictions
print('N Rooms   Median Value')
for i in range(4, 10):
    print(str(i) + '         $' + str(round(clf.predict(i)[0] * 1000, 2)))
```
